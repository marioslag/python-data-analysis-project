```python
from sympy import Symbol, diff
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
import autograd.numpy as np
from autograd import grad

x1 = sym.Symbol('x1')
x2 = sym.Symbol('x2')
def f1(x1, x2):
    return (x1 - 2)**2 + (x2 - 3)**2

def f2(x1, x2):
    return (1 - (x2 - 3)) ** 2 + 20 * ((x1 + 3) - (x2 - 3) ** 2) ** 2
# Yπολογισμος μερικης παραγωγου για να υπολογιστει το gradient
f_1 = (x1 - 2)**2 + (x2 - 3)**2
f_2 = (1 - (x2 - 3)) ** 2 + 20 * ((x1 + 3) - (x2 - 3) ** 2) ** 2
df1_dx1 = diff(f_1, x1)
df1_dx2 = diff(f_1,x2)
df2_dx1 = diff(f_2, x1)
df2_dx2 = diff(f_2, x2)
print("df1_dx1 = ", df1_dx1, '\n')
print("df1_dx2 = ", df1_dx2, '\n')
print("df2_dx1 = ", df2_dx1, '\n')
print("df2_dx2 = ", df2_dx2, '\n')
```

```python
#define gradients
def grad_f1(x1, x2):
    return np.array([2*x1 - 4 ,2*x2 - 6])
def grad_f2(x1,x2):
    return np.array([40*x1 - 40*(x2 - 3)**2 + 120 , 2*x2 + 20*(12 - 4*x2)*(x1 - (x2 - 3)**2 + 3) - 8])


#learning rate and iritations
lr = 0.5
iterations = 10
#αρχικοποιηση τιμων
x1 = np.array([0, 0])
x2 = np.array([0, 0])
f1_val = []
f2_val = []
iteration_f1 = []
iteration_f2 = []

#gradient descent
for i in range(iterations):
    f1_val.append(f1(x1[0], x1[1]))
    gradf1 = grad_f1(x1[0], x1[1])
    iteration_f1.append(i)
    x1 = x1-lr*gradf1

print ('F1 \n')
for i in range(iterations):
  print ('Value of f1 of iteration ', i+1, ':', f1_val[i], '\n' )  

print ('F2 \n')

#same for f2
for i in range(iterations):
    f2_val.append(f2(x2[0], x2[1]))
    gradf2 = grad_f2(x2[0], x2[1])
    iteration_f2.append(i)
    x2 = x2-lr*gradf2
for i in range(iterations):
  print ('Value of f2 of iteration ', i+1, ':', f2_val[i], '\n' )  


#Plot για την τιμή των συναρτήσεων συναρτήσει των επαναλήψεων



#f1
plt.plot(iteration_f1, f1_val)
plt.xlabel("No of Iterations")
plt.ylabel("F1 value")
plt.show


```
